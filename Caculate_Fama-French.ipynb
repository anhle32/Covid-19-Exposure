```python
unique_stocks = df_result['id'].unique()
quarters = df_result['Quarter'].unique()
years = df_result['Year'].unique()
```

```python
from sklearn.linear_model import LinearRegression
list_stock = []
list_year=[]
list_quarter=[]
list_beta_1=[]
for stock in unique_stocks:
    for year in years:
        for quarter in quarters:
            try:
                mask = (df_result['id'] == stock) & (df_result['Quarter'] == quarter) & (df_result['Year'] == year)
                subset_data = df_result[mask].copy()
                y=subset_data['return']
                x=subset_data[['new_cases','return_vni','SMB']]
                model=LinearRegression()
                model.fit(x,y)
                list_stock.append(stock)
                list_year.append(year)
                list_quarter.append(quarter)
                list_beta_1.append(model.coef_[0])     
            except Exception:
                continue
```

```python
df_beta_1 = pd.DataFrame({
    'id': list_stock,
    'year': list_year,
    'quarter': list_quarter,
    'beta_1': list_beta_1
})
df_beta_1
```

```python
list_stock = []
list_year=[]
list_quarter=[]
list_beta_2=[]
for stock in unique_stocks:
    for year in years:
        for quarter in quarters:
            try:
                mask = (df_result['id'] == stock) & (df_result['Quarter'] == quarter) & (df_result['Year'] == year)
                subset_data = df_result[mask].copy()
                y=subset_data['return']
                x=subset_data[['new_cases','return_vni','SMB','HML']]
                model=LinearRegression()
                model.fit(x,y)
                list_stock.append(stock)
                list_year.append(year)
                list_quarter.append(quarter)
                list_beta_2.append(model.coef_[0])     
            except Exception:
                continue
```

```python
df_beta_2 = pd.DataFrame({
    'id': list_stock,
    'year': list_year,
    'quarter': list_quarter,
    'beta_2': list_beta_2
})
df_beta_2
```

```python
df_beta_2['beta_1']=df_beta_1['beta_1']
df_beta_2
```
